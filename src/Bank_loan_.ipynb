### Total Loan Applications

```python
total_loan_applications = df['id'].count()
print("Total Loan Applications:", total_loan_applications)
```

### MTD Total Loan Applications

```python
##Month to date application - here we have historical data and the data is till December 2022, 
##so we will consider the month of December 2022 for this analysis. 
##We will filter the data for the month of December 2022 and then count the number of loan applications received in that month.
## If you are doing any analysis for the current month, you can filter the data for the current month and 
##then count the number of loan applications received in that month.

latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data = df[(df['issue_date'].dt.year == latest_year) & (df['issue_date'].dt.month == latest_month)]

MTD_loan_applications = mtd_data['id'].count()

print(f"Month to Date Loan Applications(for {latest_issue_date.strftime('%B %Y')}): {MTD_loan_applications}")
```

### Total Funded Amount


```python
###total funded amount - here we will calculate the total funds granted to all the loan applicants.

total_funded_amount = df['loan_amount'].sum()
total_funded_amount_millions = total_funded_amount / 1000000  # Convert to millions

print(f"Total Funded Amount: ${total_funded_amount_millions:.2f}M") ## :.2f is used to format the number to 2 decimal places.
```

### MTD Total Funded Amount

```python
latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data = df[(df['issue_date'].dt.year == latest_year) & (df['issue_date'].dt.month == latest_month)]

mtd_total_funded_amount = mtd_data['loan_amount'].sum()
mtd_total_funded_amount_millions = mtd_total_funded_amount / 1000000  # Convert to millions

print(f"Month to Date Total Funded Amount: ${mtd_total_funded_amount_millions:.2f}M")
```

### Total Amount Received

```python
total_amount_received = df['total_payment'].sum()
total_amount_received_millions = total_amount_received / 1000000  # Convert to millions

print(f"Total Amount Received: ${total_amount_received_millions:.2f}M")
```

### MTD Total Amount Received

```python
latest_issue_date = df['issue_date'].max()
latest_year = latest_issue_date.year
latest_month = latest_issue_date.month

mtd_data = df[(df['issue_date'].dt.year == latest_year) & (df['issue_date'].dt.month == latest_month)]

mtd_total_amount_received = mtd_data['total_payment'].sum()
mtd_total_amount_received_millions = mtd_total_amount_received / 1000000  # Convert to millions

print(f"Month to Date Total Amount Received: ${mtd_total_amount_received_millions:.2f}M")
```

### Average Interest Rate

```python
## to find the average interest rate, we will calculate the mean of the 'int_rate' column in the dataset and
## convert it in to % we will use *100 to convert it into percentage.
average_interest_rate = df['int_rate'].mean() * 100
print(f"Average Interest Rate: {average_interest_rate:.2f}%") 
```

### Average Debt-to-Income Ratio(DTI)

```python
average_debt_to_income_ratio = df['dti'].mean() * 100
print(f"Average debt-to-income Ratio: {average_debt_to_income_ratio:.2f}%") 
```

### Good Loan Matrics

```python
###Good Loans
good_loan_applications = df[df['loan_status'].isin(['Fully Paid', 'Current'])]

total_loan_applications = df['id'].count()

good_loan_applications_count = good_loan_applications['id'].count()
good_loan_funded_amount = good_loan_applications['loan_amount'].sum()
good_loan_received_amount = good_loan_applications['total_payment'].sum()

good_loan_funded_amount_millions = good_loan_funded_amount / 1000000  # Convert to millions
good_loan_received_amount_millions = good_loan_received_amount / 1000000  # Convert to millions

good_loan_applications_percentage = (good_loan_applications_count / total_loan_applications) * 100

print(f"Good Loan Applications Count: {good_loan_applications_count}")
print(f"Good Loan Funded Amount: ${good_loan_funded_amount_millions:.2f} million")
print(f"Good Loan Received Amount: ${good_loan_received_amount_millions:.2f} million")
print(f"Good Loan Applications Percentage: {good_loan_applications_percentage:.2f}%")
```

### Bad Loans Metrics

```python
## bad Loans - here we will filter the data for the loan applications that have been charged off.
bad_loan_applications = df[df['loan_status'].isin(['Charged Off'])] 

total_loan_applications = df['id'].count()

bad_loan_applications_count = bad_loan_applications['id'].count()
bad_loan_funded_amount = bad_loan_applications['loan_amount'].sum()
bad_loan_received_amount = bad_loan_applications['total_payment'].sum()

bad_loan_funded_amount_millions = bad_loan_funded_amount / 1000000  # Convert to millions
bad_loan_received_amount_millions = bad_loan_received_amount / 1000000  # Convert to millions   

bad_loan_applications_percentage = (bad_loan_applications_count / total_loan_applications) * 100

print(f"Bad Loan Applications Count: {bad_loan_applications_count}")
print(f"Bad Loan Funded Amount: ${bad_loan_funded_amount_millions:.2f} million")
print(f"Bad Loan Received Amount: ${bad_loan_received_amount_millions:.2f} million")
print(f"Bad Loan Applications Percentage: {bad_loan_applications_percentage:.2f}%")
```

## Charts

### Monthly Trends by Issue Date for Total Funded Amount

```python
monthly_funded_amount = (df.sort_values('issue_date').assign(month_name=lambda x: x['issue_date'].dt.strftime('%Y-%m'))
                        .groupby('month_name', sort=False)['loan_amount'].sum().div(1000000).reset_index(name='loan_amount_millions'))

plt.figure(figsize=(10,5))
plt.fill_between(monthly_funded_amount['month_name'], monthly_funded_amount['loan_amount_millions'], color='skyblue', alpha=0.5)
plt.plot(monthly_funded_amount['month_name'], monthly_funded_amount['loan_amount_millions'], color='black', linewidth=2) 

## we would like to add data lables to the plot to show the exact funded amount for each month. 
# We can use the plt.text() function to add text to the plot.

for i, row in monthly_funded_amount.iterrows():
    plt.text(i, row['loan_amount_millions'] + 0.3,
             f"${row['loan_amount_millions']:.2f}M",
             ha='center', va='bottom')

plt.title('Monthly Funded Amount (in Millions)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Funded Amount ($ in Millions)', fontsize=12)  
plt.xticks(ticks=range(len(monthly_funded_amount)), labels=monthly_funded_amount['month_name'], rotation=45)
plt.scatter(monthly_funded_amount['month_name'], monthly_funded_amount['loan_amount_millions'], color='black', s=30)  # Add scatter points
plt.tight_layout()
plt.show()
```

### Monthly Trends by Issue Date for Total Amount Received

```python
monthly_received_amount = (df.sort_values('issue_date').assign(month_name=lambda x: x['issue_date'].dt.strftime('%Y-%m'))
                        .groupby('month_name', sort=False)['total_payment'].sum().div(1000000).reset_index(name='received_amount_millions'))

plt.figure(figsize=(10,5))
plt.fill_between(monthly_received_amount['month_name'], monthly_received_amount['received_amount_millions'], color='lightgreen', alpha=0.5)
plt.plot(monthly_received_amount['month_name'], monthly_received_amount['received_amount_millions'], color='green', linewidth=2) 

## we would like to add data lables. 
# We can use the plt.text() function to add text to the plot.

for i, row in monthly_received_amount.iterrows():
    plt.text(i, row['received_amount_millions'] + 0.3,
             f"${row['received_amount_millions']:.2f}M",
             ha='center', va='bottom')

plt.title('Monthly Received Amount (in Millions)', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Received Amount ($ in Millions)', fontsize=12)  
plt.xticks(ticks=range(len(monthly_received_amount)), labels=monthly_received_amount['month_name'], rotation=45)
plt.scatter(monthly_received_amount['month_name'], monthly_received_amount['received_amount_millions'], color='green', s=30)  # Add scatter points
plt.tight_layout()
plt.show()
```

### Monthly Trends by Issue Date for Total Loan Applications

```python
monthly_loan_applications = (
    df.sort_values('issue_date')
      .assign(month_name=lambda x: x['issue_date'].dt.strftime('%Y-%m'))
      .groupby('month_name', sort=False)['id'].count()
      .reset_index(name='loan_applications_count')
)

plt.figure(figsize=(10,5))
plt.fill_between(monthly_loan_applications['month_name'],
                 monthly_loan_applications['loan_applications_count'],
                 color='orange', alpha=0.5)
plt.plot(monthly_loan_applications['month_name'], monthly_loan_applications['loan_applications_count'], color='orange', linewidth=2) 

## we would like to add data lables. 
# We can use the plt.text() function to add text to the plot.

for i, row in monthly_loan_applications.iterrows():
    plt.text(i, row['loan_applications_count'] + 0.3,
             f"{row['loan_applications_count']:}",
             ha='center', va='bottom')

plt.title('Total Loan Applications by Month', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Applications', fontsize=12)  
plt.xticks(ticks=range(len(monthly_loan_applications)), labels=monthly_loan_applications['month_name'], rotation=45)
plt.scatter(monthly_loan_applications['month_name'], monthly_loan_applications['loan_applications_count'], color='orange', s=30)  # Add scatter points
plt.tight_layout()
plt.show()
```

### Regional Analysis by State by Total Funded Amount

```python
state_funded_amount = (
    df.groupby('address_state')['loan_amount']
      .sum()
      .div(1000)
      .reset_index(name='state_funded_amount_thousands')
      .rename(columns={'address_state': 'state'})
      .sort_values('state_funded_amount_thousands', ascending=True)
)

plt.figure(figsize=(10,8))
bars = plt.barh(state_funded_amount['state'], state_funded_amount['state_funded_amount_thousands'], color='skyblue', alpha=0.7)

for bar in bars:
    width = bar.get_width()
    plt.text(width + 10, bar.get_y() + bar.get_height()/2,
             f"${width:.0f}K", va='center', fontsize=7)
    
plt.title('Total Funded Amount by State ($ in Thousands)', fontsize=14)
plt.xlabel('Funded Amount ($ in Thousands)', fontsize=12)
plt.ylabel('State', fontsize=12)
plt.tight_layout()
plt.show()
```

### Loan Term Analysis by Total Funded Amount
